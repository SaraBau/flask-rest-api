```python


df = pd.read_excel('stock_list.xlsx', index_col=0).head(100)
df
```

```python
df1 = pd.DataFrame({'City': ['Bangalore', 'Pune', 'Chennai', 'Mumbai','New Delhi'],
                     'Company': ['INFY', 'TCS', 'HDFCBANK', 'HINDUNILVR','RELIANCE']})
df1
```

```python
merged_inner = pd.merge(left=df, right=df1, left_on='NSE Code', right_on='Company')
# In this case `species_id` is the only column name in  both dataframes, so if we skipped `left_on`
# And `right_on` arguments we would still get the same result

# What's the size of the output data?
merged_inner.shape
merged_inner
```

```python
merged_left = pd.merge(left=df, right=df1, how='right', left_on='NSE Code', right_on='Company')
merged_left
```
